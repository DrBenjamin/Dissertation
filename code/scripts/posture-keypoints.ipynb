# Posture Keypoints Detection with YOLO11

This notebook trains a YOLO11 model for posture keypoint detection using a custom dataset. The workflow includes:

1. **Model Setup**: Loading a pretrained YOLO11n (nano) model as the base for transfer learning
2. **Training**: Fine-tuning the model on a custom posture dataset with keypoint annotations
3. **Configuration**: 
   - Training for 200 epochs
   - Image size: 640x640 pixels
   - Batch size: 32

## Prerequisites

- Ultralytics YOLO library
- Custom posture dataset configured in `data.yaml` format
- Pretrained YOLO11n weights

## Dataset Structure

The training data should be organized according to the specifications in `../data/posture/data.yaml`, which defines:
- Training and validation image directories
- Keypoint definitions for posture analysis
- Class labels

## Model Output

Training results will be saved in `runs/detect/` including:
- Trained model weights
- Training metrics and logs
- Validation results

## Documentation

For more information on YOLO11, see: https://docs.ultralytics.com/models/yolo11/

```python
# Installing the ultralytics package
%pip install ultralytics
```

```python
from ultralytics import YOLO
model = YOLO("../data/models/yolo11n.pt")  # load a pretrained model (recommended for training)

# Training the model
results = model.train(data="../data/posture/data.yaml", epochs=200, imgsz=640, batch=32)
```
